```python
data_agency['price'].describe()
```

```python
Q1 = data_agency.price.quantile(0.25)
Q3 = data_agency['price'].quantile(0.75)
Q1,Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR

lower_limit,upper_limit
```

```python
data_outliers=data_agency[(data_agency.price<lower_limit)|(data_agency.price>upper_limit)]
print(data_outliers)
```

```python
data_outliers.shape
```

```python
data_no_outliers=data_agency[(data_agency.price>lower_limit)&(data_agency.price<upper_limit)]
print(data_no_outliers)
```

```python
data_no_outliers.shape
```

```python
new_data_agency = data_agency.copy()

for x in ['price']:
    q75,q25 = np.percentile(new_data_agency.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
    #remplacer les out par nan
    new_data_agency.loc[new_data_agency[x] < min,x] = np.nan
    new_data_agency.loc[new_data_agency[x] > max,x] = np.nan
    
```

```python
data_agency.isnull().sum()
```

```python
new_data_agency.isnull().sum()
```

```python
#imputing => missing value remplacer par mean
data_agencyN=new_data_agency.fillna(new_data_agency["price"].mean())
data_agencyN
```

```python
data_agencyN.isnull().sum()
```

```python
#Compare the plots after capping
plt.figure(figsize=(16,8))

sns.distplot(data_agencyN['price'])
plt.subplot(2,2,4)
sns.boxplot(data_agencyN['price'])
plt.show()
```
